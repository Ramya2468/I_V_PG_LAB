```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = r"D:\Mtech Docs\Second Sem\Image and Video\PG_lab\Image1.jpg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Negative Transformation
def negative_transformation(img):
    return 255 - img

# Histogram Equalization
def histogram_equalization(img):
    return cv2.equalizeHist(img)

# Power-law (Gamma) Transformation
def gamma_transformation(img, gamma):
    gamma_img = np.array(255 * (img / 255) ** gamma, dtype='uint8')
    return gamma_img

# Plot histogram function
def plot_histogram(img, title, ax):
    ax.hist(img.ravel(), 256, [0, 256])
    ax.set_title(title)
    ax.set_xlabel('Pixel Intensity')
    ax.set_ylabel('Frequency')

# Applying the transformations
negative_img = negative_transformation(img)
gamma_img = gamma_transformation(img, gamma=0.5)
equalized_img = histogram_equalization(img)

# Plotting the images and histograms
fig, axes = plt.subplots(3, 2, figsize=(8, 8))

# Original Image and Histogram
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')
plot_histogram(img, 'Original Histogram', axes[0, 1])

# Negative Transformation and Histogram
axes[1, 0].imshow(negative_img, cmap='gray')
axes[1, 0].set_title('Negative Image')
axes[1, 0].axis('off')
plot_histogram(negative_img, 'Negative Histogram', axes[1, 1])

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.show()





# Gamma Transformation and Histogram
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

# Plot Gamma Transformation Image
axes[0].imshow(gamma_img, cmap='gray')
axes[0].set_title('Gamma Image')
axes[0].axis('off')  # Turn off the axis

# Plot Histogram of Gamma Image
plot_histogram(gamma_img, 'Gamma Histogram', axes[1])

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.show()


# Plot Histogram Equalization and Histogram
fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].imshow(equalized_img, cmap='gray')
axes[0].set_title('Equalized Image')
axes[0].axis('off')
plot_histogram(equalized_img, 'Equalized Histogram', axes[1])

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.show()

# Plot Histograms Before and After Equalization
fig, axes = plt.subplots(1, 2, figsize=(8, 3))
plot_histogram(img, 'Histogram Before Equalization', axes[0])
plot_histogram(equalized_img, 'Histogram After Equalization', axes[1])

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = r'D:\Mtech Docs\Second Sem\Image and Video\PG_lab\Image1.jpg'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Linear Contrast Stretching
def linear_stretching(img):
    # Find the minimum and maximum pixel values in the image
    min_val = np.min(img)
    max_val = np.max(img)

    # Apply the linear transformation formula
    stretched_img = (img - min_val) * (255 / (max_val - min_val))

    # Convert to uint8 type
    stretched_img = np.clip(stretched_img, 0, 255).astype(np.uint8)
    return stretched_img

# Foggy Effect
def foggy_effect(img, intensity=0.5):
    # Create a foggy effect by blending the image with a white image
    fog = np.ones(img.shape, dtype=np.uint8) * 255  # White image
    foggy_img = cv2.addWeighted(img, 1 - intensity, fog, intensity, 0)
    return foggy_img

# Apply linear stretching to the image
stretched_img = linear_stretching(img)

# Apply foggy effect to the stretched image
foggy_img = foggy_effect(stretched_img, intensity=0.3)  # Adjust intensity for more/less fog

# Plotting the original, stretched, and foggy images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(stretched_img, cmap='gray')
plt.title('Stretched Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(foggy_img, cmap='gray')
plt.title('Foggy Image')
plt.axis('off')

plt.show()

# Plotting the histograms
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Original Histogram')

plt.subplot(1, 3, 2)
plt.hist(stretched_img.ravel(), 256, [0, 256])
plt.title('Stretched Histogram')

plt.subplot(1, 3, 3)
plt.hist(foggy_img.ravel(), 256, [0, 256])
plt.title('Foggy Histogram')

plt.show()
```
